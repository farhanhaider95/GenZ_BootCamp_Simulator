```python
# Welcome Message
def GreetUser():
    print("Welcome to the GenZ Bootcamp Cadet Simulator!")
# Profile Setup
user = {}
user["name"] = input("Enter your name: ")
user["rank"] = input("Enter your rank: ")
user["age"] = int(input("Enter your age: "))
print(f"Hello {user['rank']} {user['name']}! Ready to survive the bootcamp?")
# Chatbot Responses
responses = {
    1: "Welcome to the bootcamp, soldier! ‚öî",
    2: "Mission-ready and caffeinated ‚òï ",
    3: "Dismissed! See you at next drill",
    4: "Ooo! want to play Bootcamp Simulator with me? (yes/no)",
}
def ShowChatMenu():
    print("\n Choose a response:")
    print("1. Say 'hi'")
    print("2. Ask 'how are you?'")
    print("3. Say 'bye'")
    print("4. Start the game")
def GetResponse(choice):
    return responses.get(choice, "That's CLASSIFIED, I could tell you but then I'd have to ‚ò† you!")
def EndChat():
    print(f"See you in the AfterLife {user['rank']} !")
# Stats Setup
stats = {
    "energy": 100,
    "mood": "Neutral",
    "discipline": 0,
    "skill": 0,
    "hydration": 50
}
# GenZ Problems
import random
def TriggerGenZProblem():
    problems = [
        "No WiFi in the barracks ",
        "TikTok ban during training ",
        "Boots don't match the outfit ",
        "Commander doesn't understand memes ",
        "Zero time for skincare routine ",
        "Can't find time to post gym selfies üì∏ ",
        "Woke up before sunrise... again üòµ "
    ]
    problem = random.choice(problems)
    print(f"GenZ Problem: {problem}")
    stats["mood"] = "Annoyed"
    stats["energy"] -= 5
# Game Functions
def ShowStatus():
    print(f"Status -> Energy: {stats['energy']}, Mood: {stats['mood']}, Discipline: {stats['discipline']}, Skill: {stats['skill']}, Hydration: {stats['hydration']} ")
def UpdateStats(activity):
    if activity == "drill":
        stats["energy"] -= 30
        stats["discipline"] += 15
        stats["mood"] = "Focused"
        stats["hydration"] -= 20
    elif activity == "eat":
        stats["energy"] += 20
        stats["mood"] = "Satisfied"
        stats["hydration"] += 20
    elif activity == "maintain":
        stats["energy"] -= 15
        stats["skill"] += 10
        stats["mood"] = "Sharp"
        stats["hydration"] -= 10
    elif activity == "sleep":
        stats["energy"] = 100
        stats["mood"] = "Rested"
        stats["hydration"] -= 5
    elif activity == "hydrate":
        stats["energy"] += 10
        stats["hydration"] += 30
        stats["mood"] = "Refreshed"
    else:
        print("You need to be disciplined kid!")
    ShowStatus()
    TriggerGenZProblem()
def EndGame():
    print("üéñBootcamp Complete! Final Status:")
    ShowStatus()
    EndChat()
# Activity Menu
def StartGame():
    print("\n Activities:")
    print("1. Drill üí™üèº ")
    print("2. Eat at Mess Hall üî• ")
    print("3. Weapon Maintenance ü™ì")
    print("4. Sleep üõè ")
    print("5. Hydrate üíß ")
    print("6. Quit Game üí® ")
    while True:
        choice = input("Choose an activity (1-6): ")
        if choice == "1":
            UpdateStats("drill")
        elif choice == "2":
            UpdateStats("eat")
        elif choice == "3":
            UpdateStats("maintain")
        elif choice == "4":
            UpdateStats("sleep")
        elif choice == "5":
            UpdateStats("hydrate")
        elif choice == "6":
            EndGame()
            break
        else:
            print(f"You need to Reconsider your choices {user['rank']}!")
# Menu-Driven Chat Loop
GreetUser()
while True:
    ShowChatMenu()
    try:
        user_choice = int(input("Enter your choice (1-4): "))
        if user_choice == 3:
            print(GetResponse(user_choice))
            EndChat()
            break
        elif user_choice == 4:
            print(GetResponse(user_choice))
            play_choice = input("Type 'yes' to start or 'no' to continue chatting: ").lower()
            if play_choice == "yes":
                StartGame()
                break
            else:
                continue
        else:
            print("AI Instructor:", GetResponse(user_choice))
    except ValueError:
        print("Please enter a valid number.")
```
